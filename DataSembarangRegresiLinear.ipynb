Library yang digunakan:
- numpy → perhitungan numerik
- matplotlib → visualisasi data
- scikit-learn → model machine learning

Kasus:
Menganalisis hubungan antara **penjualan** dan **harga**.

```python
# Import library yang dibutuhkan
import numpy as np  # untuk perhitungan numerik
import matplotlib.pyplot as plt  # untuk visualisasi
from sklearn.linear_model import LinearRegression
```

## 📌 Membuat Data

Kita membuat data sederhana:
- Penjualan sebagai variabel X
- Harga sebagai variabel Y

Data ini hanya contoh (dummy data).

```python
# Membuat data sembarang
penjualan = np.array([6,5,5,4,4,3,2,2,2,1])
harga = np.array([16000, 18000, 27000, 34000, 50000, 68000, 65000, 81000, 85000, 90000])

# Print data
print("Data Penjualan :", penjualan)
print("Data Harga :", harga)
```

## 📊 Visualisasi Data Awal

Grafik scatter digunakan untuk melihat hubungan antara penjualan dan harga.

```python
plt.scatter(penjualan, harga)
plt.title("Scatter Plot Data Penjualan vs Harga")
plt.xlabel("Penjualan")
plt.ylabel("Harga")
plt.show()
```

## 🤖 Membuat Model Regresi Linear

Langkah:
1. Mengubah bentuk data (reshape)
2. Membuat model Linear Regression
3. Melatih model menggunakan data

```python
# Reshape data (wajib untuk sklearn)
penjualan = penjualan.reshape(-1, 1)

# Membuat model regresi
linreg = LinearRegression()
linreg.fit(penjualan, harga)
```

## 📈 Visualisasi Hasil Regresi

- Titik merah = data asli
- Garis = hasil prediksi model

```python
plt.scatter(penjualan, harga, color='red')
plt.plot(penjualan, linreg.predict(penjualan))

plt.title("Visualisasi Model Regresi Data Penjualan dan Harga")
plt.xlabel("Penjualan")
plt.ylabel("Harga")

plt.show()
```

## 🧠 Insight

- Semakin tinggi harga, penjualan cenderung menurun
- Model berhasil membuat garis hubungan linear
- Ini contoh sederhana supervised learning

Kesimpulan:
Regresi linear dapat digunakan untuk memprediksi hubungan antar variabel numerik.

```python
# Prediksi contoh
prediksi = linreg.predict([[3]])
print("Prediksi harga untuk penjualan 3:", prediksi)
```
